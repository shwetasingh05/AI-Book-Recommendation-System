```python
#create a clean dataset 
df = df = df[['book_id','title','authors','publisher','published_date','categories','description',
         'average_rating','ratings_count','page_count','language']]
```

```python
df = df.dropna(subset=['title','authors','publisher','categories','description'])
```

```python
df['average_rating'] = df['average_rating'].fillna(df['average_rating'].mean())
```

```python
df = df.drop_duplicates(subset='title')
```

```python
df = df.reset_index(drop=True)
```

```python
#In columns publisher 
#remove missing publishers 
df['publisher'] = df['publisher'].fillna('Unknown')
#remove extra spaces 
df['publisher'] = df['publisher'].str.strip()
#convert to consistent format
df['publisher'] = df['publisher'].str.title()
```

```python
df['published_date'] = pd.to_datetime(df['published_date'], utc=True)
#extracting year
df['published_year'] = df['published_date'].dt.year
#extracting month
df['published_month'] = df['published_date'].dt.month
#removing invalid years
df = df[df['published_year'] > 1900]
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
#Remove Rows with Missing book_id
df = df.dropna(subset=['book_id'])
```

```python
#Fill Missing Page Count
df['page_count'] = df['page_count'].fillna(df['page_count'].median())
```

# FEATURE ENGINEERING

```python
#When multiple important "text columns" are combined together into "one" columns then it is called FEATURE.
#Machine Learning compares text information to understand similarity between books.
#If columns are kept in different colummns, the algorithm will not be able to compare them easily.
```

```python
df['features'] = df[['title', 'authors', 'categories', 'publisher', 'description']].astype(str).agg(' '.join, axis=1)
```

```python
#convert to lower case
df['features'] = df['features'].str.lower()
```

```python
df[['title','features']].head()
```

```python
#Convert Text into Numbers (TF-IDF)
#TF-IDF is a technique used to convert text into numbers so that machines can understand and analyze text data.


from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['features'])
```

```python
tfidf_matrix.shape
```

### Similar Book recommendation

```python
#calculating book similarity
#Calculate how 2 books are similar
#METHOD: Cosine Similarity - compares two vectors

from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
similarity_matrix.shape
```

```python
# Build the Recommendation Function
def recommend_similar(book_title):

    book_title = book_title.lower()

    matches = df[df['title'].str.contains(book_title, case=False, na=False)]

    if matches.empty:
        print("Book not found in dataset.")
        return

    index = matches.index[0]

    similarity_scores = list(enumerate(similarity_matrix[index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:6]

    for i in similarity_scores:
        book = df.iloc[i[0]]
        print(book['title'], "-", book['authors'], "(Rating:", book['average_rating'], ")")
```

```python
df['title'].head(10)
```

```python
recommend_similar("bestsellers")
```

```python
recommend_similar("Young Adults")
```

### Popular Book recommendation

```python
#Popularity-Based Recommendation System
#Popularity Score Column
df['popularity_score'] = df['average_rating'] * np.log(df['ratings_count'] + 1)
```

```python
#Sort Books by Popularity
popular_books = df.sort_values(by='popularity_score', ascending=False)
```

```python
#View Top Popular Books
popular_books[['title','authors','average_rating','ratings_count','popularity_score']].head(10)
```

```python
#Create Popular Recommendation Function
def recommend_popular(n=10):
    
    top_books = df.sort_values(by='popularity_score', ascending=False).head(n)
    
    for _, book in top_books.iterrows():
        print(book['title'], "-", book['authors'], "(Rating:", book['average_rating'], ")")
```

```python
recommend_popular()
```

### HYBRID RECOMMENDATION SYSTEM : Similar Books + Most Popular among similar books

```python
def recommend_hybrid(book_title, n=5):

    book_title = book_title.lower()

    matches = df[df['title'].str.contains(book_title, case=False, na=False)]

    if matches.empty:
        print("Book not found in dataset.")
        return

    index = matches.index[0]

    similarity_scores = list(enumerate(similarity_matrix[index]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similar_books_index = [i[0] for i in similarity_scores[1:50]]

    similar_books = df.iloc[similar_books_index]

    ranked_books = similar_books.sort_values(by='popularity_score', ascending=False)

    for _, book in ranked_books.head(n).iterrows():
        print(book['title'], "-", book['authors'], "(Rating:", book['average_rating'], ")")
```

```python
recommend_hybrid("game of thrones")
```
